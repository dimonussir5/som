```
import numpy as np
import pandas as pd
from sklearn import datasets
```

```
# load iris dataset
iris = datasets.load_iris()
# convert to dataframe (Just for easy access) 
iris_dataset = pd.DataFrame(iris.data)
# load iris labels 
iris_labels = iris.target

iris_dataset.head()
```

```
# less use standary scaler on the data
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()
iris_dataset_standar = pd.DataFrame(standard.fit_transform(iris_dataset))
iris_dataset_standar.head()
```

```
from som import SOM

# initialize the SOM
som = SOM(15, 15)

# fit the SOM for 20000 epochs, save the error every 300 steps
som.fit(iris_dataset_standar.to_numpy(), 20000, save_e=True, interval=300) 
```

```
# plot the training error history
som.plot_error_history() 
```

```
som.plot_point_map(iris_dataset_standar.to_numpy(), iris_labels, np.unique(iris_labels))
```
